```python
import numpy as np

M_list = [8, 16, 24, 32, 40, 48, 56, 64]
python_times = [
    0.115291777998209, 0.1767352670431137, 0.31298348493874073, 0.4145647846162319,
    0.6246098354458809, 0.843100979924202, 1.1178868003189564, 1.3279154561460018
]

julia_times = [0.128978, 0.374159, 0.668243, 1.965, 2.741, 3.712, 4.735, 6.220]


complexity = [m*np.log(m) for m in M_list]


import matplotlib.pyplot as plt

ax = plt.axes()

ax.set_yscale('log')
ax.set_xscale('log')

ax.plot(M_list, python_times)
ax.plot(M_list, complexity)
ax.plot(M_list, julia_times)
```
